```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data = np.loadtxt('pulsos26-3.txt',dtype=np.str)
k = len(data[:,1])
```

```python
tiempo = []
voltaje = []
for i in range (k):
    tiempo += [int( data[i,1])]
    voltaje += [5*int( data[i,2])/1024.0 ]
```

```python
plt.figure(figsize=(15,8))
plt.plot(tiempo, voltaje,'.')
plt.ylabel('Voltaje (V)')
plt.xlabel('Tiempo (seg)')
#plt.xlim(2800,3100)
plt.grid()
```
