```
import numpy as np
import matplotlib.pyplot as plt
```

```
datos = np.genfromtxt('PRU1710.TXT')
```

```
datos[:,0]
```

```
plt.figure(figsize=(15,9))
for i in range(15): #[2]:#
  plt.plot(datos[:,0],datos[:,i+1], '.', label='S'+str(i))
  plt.grid()
plt.legend()
plt.show()
```
